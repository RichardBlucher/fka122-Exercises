```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('data.csv', header=None)
data_array = data.to_numpy()
time = np.linspace(0, len(data_array[:,0])*0.0001, len(data_array[:,0]))
#plt.plot(time, data_array[:,0], label='Position 1')
#plt.plot(time, data_array[:,1], label='Position 2')
#plt.plot(time, data_array[:,2], label='Position 3')
plt.plot(time, data_array[:,3], label='Velocity 1')
plt.plot(time, data_array[:,4], label='Velocity 2')
plt.plot(time, data_array[:,5], label='Velocity 3')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Positions and Velocities over Time')
plt.xlim(0, 0.5)
plt.legend()
plt.show()
```

```python
import scipy.signal as signal


power = pd.read_csv('power.csv', header=None)
power_array = power.to_numpy()
freq=power_array[:,0]
print(len(freq))
plt.plot(freq, power_array[:,1], label='Particle 1')
plt.plot(freq, power_array[:,2], label='Particle 2')
plt.plot(freq, power_array[:,3], "--", label='Particle 3')
plt.ylabel('Power')
plt.xlabel('Frequency')
plt.xlim(-500, 500)
plt.title('Power Spectrum')
plt.legend()
plt.show()

peaks1 = signal.find_peaks(power_array[:,1])
peaks2 = signal.find_peaks(power_array[:,2])
peaks3 = signal.find_peaks(power_array[:,3])

print(freq[peaks1[0]])
print(freq[peaks2[0]])
print(freq[peaks3[0]])
```
