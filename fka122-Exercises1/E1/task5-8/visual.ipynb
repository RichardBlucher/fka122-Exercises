```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('energies-task8.csv', header=None)
data_array = data.to_numpy()
N = data_array.shape
print(N)
print(np.max(data_array[:,0]))
print(np.min(data_array[:,0]))
time = np.linspace(0, len(data_array[:,0])*0.1, len(data_array[:,0]))
plt.plot(time, data_array[:,0], label='Mode 1')
plt.plot(time, data_array[:,1], label='Mode 2')
plt.plot(time, data_array[:,2], label='Mode 3')
plt.plot(time, data_array[:,3], label='Mode 4')
plt.plot(time, data_array[:,4], label='Mode 5')
plt.xlabel('Time')

plt.ylabel('Energy')
plt.title('Normal Mode Energies over Time')
plt.legend()
plt.show()


```
