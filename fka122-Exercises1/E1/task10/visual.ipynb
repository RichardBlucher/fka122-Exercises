```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('cumulative_energy_task10_alpha1.csv', header=None)
data_array = data.to_numpy()
N = data_array.shape
print(N)
time = np.linspace(0, len(data_array[0,:])*100, len(data_array[0,:]))
print(time.shape)
for i in range(N[0]):
    data_array[i,:] = data_array[i,:]/time
for i in range(N[0]):
    plt.loglog(time, data_array[i,:], label=f'Mode {i+1}')




plt.xlabel('Log Time')

plt.ylabel('Log Time averged Normal Mode Energy')
plt.ylim(1e-5, 1)
plt.title('Normal Mode Energies over Time (alpha = 0.1)')
#plt.legend()
plt.show()


```
